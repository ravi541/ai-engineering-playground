```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc= nlp("Data sciene and ai & gen ai has great career ahead")
```

```python
doc
```

```python
for token in doc:
    print(token.pos_)
```

```python
for token in doc:
    print(token.text,':',token.pos_)
```

```python
for token in doc:
    print(token.text,':',token.pos_,'-->',token.lemma_,token.dep_)
```

```python
for token in doc:
    print(token.text,':',token.pos_,'-->',token.lemma_,token.dep_, token.tag_,token.shape_,token.is_alpha,token.is_stop)
```

```python
for token in doc:
    print(
        f"""
Word       : {token.text}
POS        : {token.pos_}
Lemma      : {token.lemma_}
Dependency : {token.dep_}
Tag        : {token.tag_}
Shape      : {token.shape_}
Is Alpha   : {token.is_alpha}
Is Stopword: {token.is_stop}
------------------------------
"""
    )
```

```python
for token in doc:
    print(
        f"{token.text:10} | "
        f"POS: {token.pos_:6} | "
        f"Lemma: {token.lemma_:10} | "
        f"Dep: {token.dep_:8} | "
        f"Tag: {token.tag_:5} | "
        f"Shape: {token.shape_:8} | "
        f"Alpha: {token.is_alpha} | "
        f"Stopword: {token.is_stop}"
    )
```
