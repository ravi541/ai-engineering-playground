```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.metrics import r2_score

```

```python
data = pd.read_csv(r"C:\Users\Raviteja\Downloads\car-mpg.csv")
data.head()
```

```python
data.describe()
```

```python
data.columns
```

```python
data=data.drop(['car_name'],axis=1)
```

```python
data
```

```python
data['origin']=data['origin'].replace({1:'america',2:'europe',3:'asia'})
```

```python
data
```

```python
data=pd.get_dummies(data,columns=['origin'],dtype=int)
```

```python
data = data.replace('?',np.nan)
```

```python
data = data.apply(pd.to_numeric, errors='coerce')
```

```python
#fill median only for numeric values
data=data.apply(lambda x:x.fillna(x.median()) if x.dtype != 'object' else x,axis=0)
```

```python
data.head()
```

```python
x=data.drop( ['mpg'],axis=1)
y=data[['mpg']]
```

```python
x.head()
```

```python
y.head()
```

```python
x_s=preprocessing.scale(x)
```

```python
x_s
```

```python
y_s=preprocessing.scale(y)
y_s
```

```python
x_s=pd.DataFrame(x_s,columns=x.columns)
y_s=pd.DataFrame(y_s,columns=y.columns)
```

```python
x_s
```

```python
y_s
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_s,y_s,test_size=0.3,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
regression_model = LinearRegression()
regression_model.fit(x_train,y_train)

for idx,col_name in enumerate(x_train.columns):
    print('The coef for {} is {}'.format(col_name,regression_model.coef_[0][idx]))
    
print('======')

intercept = regression_model.intercept_[0]    
print('The intercept is {}'.format(intercept))

print('======')
```
